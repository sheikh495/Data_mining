```python
data['culmen_length_mm'].fillna((data['culmen_depth_mm'].median()), inplace=True)
data['culmen_depth_mm'].fillna((data['culmen_depth_mm'].median()), inplace=True)
data['flipper_length_mm'].fillna((data['flipper_length_mm'].median()), inplace=True)
data['body_mass_g'].fillna((data['body_mass_g'].median()), inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_columns = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']
y_columns = ['species']
```

```python
x = data[x_columns].values
y = data[y_columns].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)
```

```python
from sklearn.linear_model import LogisticRegression
params = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
for i in params:
    LR = LogisticRegression(C = i)
    LR.fit(X_train, y_train)
    print("The prediction accuracy is for C equal to"+" "+str(i)+"is"+" "+str(round(LR.score(X_test, y_test), 3)))
```
